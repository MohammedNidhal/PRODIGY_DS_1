```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # To make the visulations better with more style
```

```python
df_MA=pd.read_csv("POP_MA.csv")
df_MA_Meta=pd.read_csv("Meta_MA.csv")
df_MA
```

```python
df_MA.columns.values # Checking the columns names and making them more descriptive
```

```python
df_FE=pd.read_csv("POP_FE.csv")
```

```python
df_MA.columns=list(df_FE.columns.values)[:-1] # used the Population female csv to correct the comumns names 
```

```python
df_MA
```

```python
df_MA["Region"]=df_MA_Meta["Region"] #Used the region 
```

```python
df_MA
```

```python
df_MA.isnull().sum() #Check for nulll  values in the data
```

```python
df_MA=df_MA[df_MA["Country Code"] != "INX"] # Removed the raw with all null values 
```

```python
df_MA.isnull().sum()
```

```python
df_MA_Countries = df_MA[pd.notna(df_MA['Region'])] # Extracted the countries pouplition and exclduded the Economic groups, Geographical groups etc
```

```python
df_MA_Countries
```

```python
df_MA_Countries.reset_index(drop=True, inplace=True) #Reset index to have more organazied structure
```

```python
df_MA_Countries
```

```python
df_MA_Countries.describe() # Description of the data set in hand and get more knowladge and clear  vision of the whole data
```

```python
sns.histplot(df_MA_Countries['1996'],bins=10,kde=True)
plt.title("Male Distrubition around the world on 1996 ")
plt.xlabel("Population")
plt.ylabel("Frequency") 
#Histograph to show the Population around the world and it's frequency 
```

```python
df_MA_Countries.groupby("Region")['1996'].mean().sort_values(ascending=True)
```

```python
Last_6_Male_Countries=df_MA_Countries.sort_values(by="2010",ascending=True).head(6)
sns.barplot(x="2010", y="Country Name",data=Last_6_Male_Countries)
plt.title("Top 6 countries with the least number of men")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
#Used bar chart to know what countries have the least number of men in it's demographical Population
```

```python
Top_15_Male_Countries=df_MA_Countries.sort_values(by="2015",ascending=False).head(15)
sns.barplot(x="2015", y="Country Name",data=Top_15_Male_Countries)
plt.title("Top 10 countries with men in 2015")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```

```python
df_FE=df_FE[df_FE["Country Code"] != "INX"]
```

```python
df_FE=df_FE.drop(['Country Code', 'Indicator Name', 'Indicator Code'],axis=1)
```

```python
df_FE=df_FE.drop(['2023'],axis=1)
```

```python
df_FE
```

```python
df_FE_Meta=pd.read_csv("Meta_FE.csv")
df_FE_Meta
```

```python
df_FE["Region"]=df_FE_Meta["Region"]
```

```python
df_FE.sample()
```

```python
df_FE_Countries = df_FE[pd.notna(df_FE['Region'])]
```

```python
df_FE_Countries.sample()
```

```python
df_FE_Countries.reset_index(drop=True, inplace=True)
```

```python
df_FE_Countries
```

```python
Top_15_Female_Countries=df_FE_Countries.sort_values(by="2015",ascending=False).head(15)
Top_15_Female_Countries['Gender'] = 'Female'
Top_15_Male_Countries['Gender'] = 'Male'
Top_15_World_MandF= pd.concat([Top_15_Female_Countries, Top_15_Male_Countries]) #Merging the two datasets into one data set to compare the repartion of males and females in the most populeted countru
plt.figure(figsize=(12, 8))
sns.barplot(x="2015", y="Country Name", hue="Gender", data=Top_15_World_MandF)
plt.title("Top 15 Countries by Population in 2015 (Male vs Female)")
plt.xlabel("Population")
plt.ylabel("Country")
plt.legend(title="Gender")
plt.show()
```
